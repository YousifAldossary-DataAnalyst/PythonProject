In this Project we will be writing a code to compute Newton's method for interpolant polynomials based on the given information in the assignment. However, newtowns mehtod works as shown below mathematically<br>
<br>
    $$PN(x)=a0+a1(x−x0)+a2(x−x0)(x−x1)+⋯+aN(x−x0)⋯(x−xN−1)$$ <br>
    
In this question we will introduce 4 new functions
    * polyval: applying P(x)
    * polyfit: fitting data
    * CubicSpline: evaluate the function
    * newtinterp: evaluate the interpolating polynomial
We will be focusing on the function $sin(x)$ and range of $(0, 2pi)$

```python
## Q1.part(c) Using polyval, polyfit and CubicSpline
from numpy import *
from matplotlib.pyplot import *
from scipy.interpolate import *

def evalPoly(a,xdata,x):
    n = len(xdata) - 1 #Degree of polynomial
    p = a[n]
    for k in range(1,n+1):
        p = a[n-k] + (x - xdata[n-k])*p
    return p

## finding the coefficient from Newtowns methods
def coeffs(xdata, ydata):
    m = len(xdata)
    a = ydata.copy()
    for k in range(1,m):
        a[k:m] = (a[k:m] - a[k-1])/(xdata[k:m] - xdata[k-1])
    return a

x = linspace(0 ,2*pi ,5)
y = sin(x)

xdata = linspace(-1, 1, 101)
ydata = sin(xdata)

xx=linspace(0 ,2*pi ,101)
yy=sin(xx)
cs = CubicSpline(x,y)
z1 = cs(xx)
m = len(x)
n = m-1

P = polyfit(x,y,n)
z = polyval(P,xx)

plot(x, y, 'o', label = 'interpolation points')
plot(xx , yy , label ='sin (x)')
plot(xx , z1 , 'r', label ='cubic spline (not -a- knot )')
legend( loc ='lower left' , ncol =1)
show ()

plot(x,y,'o',xx,yy,xx,z,'r')
show()
```
