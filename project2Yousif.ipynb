this is a follow up to the previous question however, we will use the dervative of a function to obtain the closest value to the intersection between the function and its dervative. This is called Newton's method that is more accurate solution.
$$ x_n+1 = x_n - F(x)/F'(x) $$

```python
from numpy import *

def fixedpoint(g, x0, tol = 1.e-6, Nmax = 50):
    err = 1.0
    iteration = 0    
    xk = x0
    while (err > tol):
        iteration = iteration + 1
        if iteration >= Nmax:
            print('Failure: Algorithm fail to converge using only NMAX iterations')
            break
        err = xk
        xk = g(xk)
        err = abs(err - xk)
        print(iteration, xk, err)
    return xk

def f(x):
    y = log(x)+x
    return y

def g(x):
    y1 = 1/x + 1
    return y1

tol = 1.e-4
x0 = 1.0
x = fixedpoint(g, x0, tol)
print('The aproximate solution is: ', x)
print('And the error is: ', f(x))
```
